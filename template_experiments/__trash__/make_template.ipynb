### gender-occupation template

hypo: either {$name1} or {$name2} is a {$occupation}.
text: {$name} is a {$occupation}

```python
import pandas as pd
with open('template/template-name.csv') as f:
    male_name = []
    female_name = []
    for line in f.readlines():
        line = line.split(',')
        if line[1]=='0':
            male_name.append(line[0])
        elif line[1]=='1':
            female_name.append(line[0])
with open('template/template-occupation.csv') as f:
    occupation = []
    for line in f.readlines():
        occupation.append(line.split(',')[0].strip())
```

```python
print(len(male_name))
male_name
```

```python
female_name = female_name[1:]
print(len(female_name))
female_name
```

```python
occupation = occupation[1:]
len(occupation)
```

```python
sents = []
for o in occupation:
    for m in male_name:
        for f in female_name:
            text = f"{m} is a {o}."
            hypo = f"either {m} or {f} is a {o}."
            sents.append(f"{text}\t{hypo}\n")

            text = f"{f} is a {o}."
            hypo = f"either {m} or {f} is a {o}."
            sents.append(f"{text}\t{hypo}\n")
```

```python
sample_sents = sents[:500]
with open('sample_gender-occupation.txt', 'w') as fw:
    for s in sample_sents:
        fw.write(s)
```
