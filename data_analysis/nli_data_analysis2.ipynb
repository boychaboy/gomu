### Count names in MNLI

```python
def count_name(data, names):
    name_count = dict()
    for n in names:
        name_count[n['name']] = 0
    for d in data:
        for n in names:
            if n['name'] in d:
                name_count[n['name']] = name_count.get(n['name'], 0) + 1
    return name_count
```

```python
total_name_mnli = count_name(mnli_data, total_names)
```

```python
total_name_snli = count_name(snli_data, total_names)
```

```python
total_name_count = {}
for name in total_name_mnli.keys():
    total_name_count[name] = total_name_mnli[name] + total_name_snli[name]
total_name_count
```

```python
for name in total_names:
    name['count'] = total_name_count[name['name']]
```

## 2. Gomu occupations in NLI dataset

### Count occupation in MNLI

```python
def count_entity(entities, data):
    count = dict()
    sent = []
    for e in entities:
        count[e] = count.get(e, 0)
    for d in data:
        for e in entities:
            if e in d:
                count[e] += 1
                sent.append(d)
    return count, sent
```

```python
occurence, sentence = count_entity(occupations, mnli_data)
```

```python
occupation_mnli_occurence = sorted(occurence.items(), key=(lambda x:x[1]), reverse=True)
json.dump(occupation_mnli_occurence, open("occupation_mnli_occurence.json", 'w'))
with open("occupation_mnli_sentence.txt", 'w') as f:
    for sent in sentence:
        f.write(f"{sent}\n")
```

```python
from transformers import pipeline
```

```python
ner = pipeline("ner", grouped_entities=True)
```

```python
result = ner("Younghoon's lives in Hawaii")
result
```

```python
def count_names(data, ner_model):
    names = {}
    for d in data:
        result = ner(d)
        if len(result) > 0:
            for entity in result:
                if entity['entity_group'] == 'PER':
                    name = entity['word']
                    names[name] = names.get(name, 0) + 1
#                     print(name)
    return names
```

```python
name_mnli_count = count_names(mnli_data, ner)
```

# Gomu occupations in NLI dataset

```python
def count_occupation(data, occupations):
    sents = []
    occs = {}
    for d in data:
        for o in occupations:
            if o in d:
                sents.append(d)
                occs[o] = occs.get(o, 0) + 1
    return occs, sents
```
