```python
# load mnli train data
with open('../data/MNLI/original/multinli_1.0_train.jsonl') as f:
    json_list = list(f)
    mnli_train = []
    for json_str in json_list:
        result = json.loads(json_str)
        mnli_train.append(result)
len(mnli_train)
```

```python
mnli_data = []
SEP = ' [SEP] '
for m in mnli_train:
    sent = m['sentence1'] + SEP + m['sentence2']
    mnli_data.append(sent)
mnli_data[0]
```

```python
# load snli train data
with open('../data/SNLI/snli_1.0_train.jsonl') as f:
    json_list = list(f)
    snli_train = []
    for json_str in json_list:
        result = json.loads(json_str)
        snli_train.append(result)
len(snli_train)
```

```python
snli_data = []
SEP = ' [SEP] '
for m in snli_train:
    sent = m['sentence1'] + SEP + m['sentence2']
    snli_data.append(sent)
snli_data[0]
```

### Count names in MNLI

```python
name_count = dict()
for n in names:
    name_count[n] = name_count.get(n, 0)
for sent in mnli_data:
    for n in names:
        if n in sent:
            name_count[n] += 1
name_count
```

```python
name_mnli_occurence = sorted(name_count.items(), key=(lambda x:x[1]), reverse=True)
json.dump(name_mnli_occurence, open("name_mnli_occurence.json", 'w'))
```

## 2. Gomu occupations in NLI dataset

### Count occupation in MNLI

```python
def count_entity(entities, data):
    count = dict()
    sent = []
    for e in entities:
        count[e] = count.get(e, 0)
    for d in data:
        for e in entities:
            if e in d:
                count[e] += 1
                sent.append(d)
    return count, sent
```

```python
occurence, sentence = count_entity(occupations, mnli_data)
```

```python
occupation_mnli_occurence = sorted(occurence.items(), key=(lambda x:x[1]), reverse=True)
json.dump(occupation_mnli_occurence, open("occupation_mnli_occurence.json", 'w'))
with open("occupation_mnli_sentence.txt", 'w') as f:
    for sent in sentence:
        f.write(f"{sent}\n")
```

```python
def count_names(data, ner_model):
    names = {}
    for d in data:
        result = ner(d)
        if len(result) > 0:
            for entity in result:
                if entity['entity_group'] == 'PER':
                    name = entity['word']
                    names[name] = names.get(name, 0) + 1
#                     print(name)
    return names
```

# Gomu occupations in NLI dataset

```python
def count_occupation(data, occupations):
    sents = []
    occs = {}
    for d in data:
        for o in occupations:
            if o in d:
                sents.append(d)
                occs[o] = occs.get(o, 0) + 1
    return occs, sents
```
