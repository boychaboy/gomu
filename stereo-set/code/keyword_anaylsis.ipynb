### Load stereoset data

```python
import dataloader
```

```python
file_path = '../data/dev.json'
stereoset = dataloader.StereoSet(file_path)
```

```python
intras = stereoset.intrasentence_examples
inters = stereoset.intersentence_examples
```

```python
total_sentences = []
```

```python
for intra in intras:
    print(intra.ID)
    print(intra.bias_type)
    print(intra.target)
    print(intra.sentences[0].sentence)
    print(intra.sentences[1].sentence)
    print(intra.sentences[2].sentence)
    total_sentences.append(intra.sentences[0].sentence)
    total_sentences.append(intra.sentences[1].sentence)
    total_sentences.append(intra.sentences[2].sentence)
    print()
```

```python
for inter in inters:
    print(inter.ID)
    print(inter.bias_type)
    print(inter.target)
    print(inter.sentences[0].sentence)
    print(inter.sentences[1].sentence)
    print(inter.sentences[2].sentence)
    total_sentences.append(inter.sentences[0].sentence)
    total_sentences.append(inter.sentences[1].sentence)
    total_sentences.append(inter.sentences[2].sentence)
    print()
```

### name 

```python
from pororo import Pororo
ner = Pororo(task="ner", lang="en")
```

```python
names = []
for sent in total_sentences:
    for token in ner(sent):
        if token[1]=='PERSON':
            names.append(token[0])
```
